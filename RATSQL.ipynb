```python
from google.colab import files
import os
import sqlite3
import json
database_name = input('What name would you like to have for your database? ')

if database_name == '':
    database_name = 'new_db'

path = '/content/drive/MyDrive/RatSql-Colab/rat-sql' # change this
db_path = '/data/spider/user_database/'
database_path = path + db_path + database_name

if not os.path.exists(database_path):
    os.mkdir(database_path)
else:
    #print(sum(1 for file_name in os.listdir(path+db_path) if database_name in file_name))
    database_path += '_{}'.format(sum(1 for file_name in os.listdir(path+db_path) if database_name in file_name) + 1)
    os.mkdir(database_path)

os.chdir(database_path)
upload_schema = files.upload()
schema_file_name = list(upload_schema.keys())[0]
if schema_file_name.split('.')[0] != 'schema':
    os.rename(database_path + '/' + schema_file_name, database_path + '/schema.sql')
    schema_file_name = 'schema.sql'

with open(schema_file_name, 'r') as file:
    sql = file.read()

conn = sqlite3.connect('{}.sqlite'.format(database_name))
c = conn.cursor()

c.executescript(sql)

conn.commit()
conn.close()

os.chdir(path)

with open(path + '/get_tables.py', 'r') as f: # change this
    get_tables = f.read()
exec(get_tables)

dev = [
    {
        "db_id": database_name,
        "query": "",
        "query_toks": [],
        "query_toks_no_value": [],
        "question": "",
        "question_toks": [],
        "sql": {
            "from": {
                "table_units": [],
                "conds": []
            },
            "select": [],
            "where": [],
            "groupBy": [],
            "having": [],
            "orderBy": [],
            "limit": None,
            "intersect": None,
            "union": None,
            "except": None
        }
    }
]

with open('data/spider/new_dev.json', 'wt') as out:
    json.dump(dev, out, sort_keys=True, indent=2, separators=(',', ': '))
```
